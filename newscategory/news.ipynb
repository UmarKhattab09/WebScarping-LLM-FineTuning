```python
from pyvis.network import Network
import pandas as pd

df = pd.read_csv("../outputs/6.csv")
import networkx as nx
```

```python
df
```

```python
df= df.groupby('NewsType')['News'].count().reset_index()
```

```python
df
```

```python
G = nx.from_pandas_edgelist(
    df,
    source='News',
    target='NewsType',
    # edge_attr='News',
    create_using=nx.Graph()

)
net=Network(notebook=True,width="1000px",height="700px",font_color="black",cdn_resources='remote')
node_degree =dict(G.degree)
net.from_nx(G)
net.show("test.html")
```

```python
net.add_node("News", label="News", size=40, color="red")
```
