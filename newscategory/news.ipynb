```python
from pyvis.network import Network
import pandas as pd

df = pd.read_csv("../outputs/5_iMPACT.csv")
import networkx as nx
```

```python
df
```

```python
df =df.groupby('Impact')['News'].count().reset_index()
```

```python
df
```

```python
G = nx.from_pandas_edgelist(
    df,
    source='Impact',
    target='Impact',  # Self-loop just for structure (or you can skip this)
    edge_attr='News',
    create_using=nx.Graph()

)
net=Network(notebook=True,width="1000px",height="700px",font_color="black",cdn_resources='remote')
node_degree =dict(G.degree)
net.from_nx(G)
net.show("test.html")
```

```python
from pyvis.network import Network
import networkx as nx
import pandas as pd

# Your DataFrame
    


# Create graph
G = nx.Graph()

# Add central node
G.add_node('News', size=20)  # Adjust central size as needed

# Add edges from 'News' to each impact level with size based on count
for index, row in df.iterrows():
    impact = row['Impact']
    count = row['News']
    G.add_node(impact, size=count * 5)  # Multiply to make size more visible
    G.add_edge('News', impact)

# Visualize using Pyvis
net = Network(notebook=True, height="450px", width="100%", bgcolor="#222222", font_color="white")
net.from_nx(G)

# Set node sizes in pyvis
for node in net.nodes:
    node['value'] = G.nodes[node['id']]['size']

# Show graph
net.show("test.html")
```
